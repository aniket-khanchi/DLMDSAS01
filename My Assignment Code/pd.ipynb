Task 1: Basic Probabilities and Visualizations

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson
```

```python
k = np.arange(1, 30)

print(k)
```

```python
pmf = poisson.pmf(k, mu=16)
pmf = np.round(pmf, 4)

for val, prob in zip(k,pmf):
    print(f"k-value {val} has probability = {prob}")
```

```python
plt.plot(k, pmf, marker='o')
plt.xlabel('k')
plt.ylabel('Probability')

plt.show()
```

```python
import matplotlib.pyplot as plt

# lm = 16

def plot_mean_poisson(lm):
    mean = lm
    plt.plot([mean, mean], [0, 1], 'k--', lw=2)
    plt.text(mean + 0.1, 0.5, 'Mean: {:.2f}'.format(mean))
    plt.show()
```

```python
plot_mean_poisson(16)
```
